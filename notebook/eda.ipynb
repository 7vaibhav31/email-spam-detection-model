REMOVING EXTRA USELESS VALUE WHICH IS {"mode":"full"}

```python
df = df[df['Category'] != '{"mode":"full"']
```

```python
df.shape
```

```python
df['length'] = df['Message'].apply(len)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.to_csv("../data/email_cleaned.csv", index=False)
```

```python
df["length"].mean()
```

```python
sns.boxplot(data=df, x='Category', y='length')
plt.xlabel('Category')
plt.ylabel('Message Length')
plt.title('Relation between Message Length and Category')
plt.show()
```
